Developing equations

```python
x1 = 1
y1 = 7
x2 = 2
y2 = 10

# the slope is 'm'
m = (y2 - y1) / (x2 - x1)

# The y intercept is 'b'
b = y1 - m*x1

# The full equation
print ("y = mx + b")
print("y =", m, "x +", b)
```

Display the graph

```python
import matplotlib.pyplot as plt

x1 = 1
y1 = 7
x2 = 2
y2 = 10

# Develop equation
m = (y2 - y1) / (x2 - x1)
b = y1 - m*x1
print("y =", m, "x +", b)

# Create graph
xmin = -10
xmax = 10
ymin = -10
ymax = 10

# For the line on the graph
y3 = m*xmin + b
y4 = m*xmax + b

# Graph Setup
fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'b') # blue x axis
plt.plot([0,0],[ymin,ymax],'b') # blue y axis

# Plot linear function as a red line
plt.plot([xmin,xmax],[y3,y4],'r')

plt.show()
```

*   A town's population increases at a constant rate. In 20210 the population was 55,000. By 2012 the population had increased to 76,000. If this trend continues, predict the population in 2016.



```python
import matplotlib.pyplot as plt

x1 = 0
y1 = 55
x2 = 2
y2 = 76

# Develop equation
m = (y2 - y1) / (x2 - x1)
b = y1 - m*x1
print("y =", m, "x +", b)

# Create graph
xmin = 0
xmax = 10
ymin = 0
ymax = 150

# For the line on the graph
y3 = m*xmin + b
y4 = m*xmax + b

# Graph Setup
fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'b') # blue x axis
plt.plot([0,0],[ymin,ymax],'b') # blue y axis

# Plot linear function as a red line
plt.plot([xmin,xmax],[y3,y4],'r')

plt.grid(True)
ax.set_xlabel("time")
ax.set_ylabel("population")

# plt.legend()
plt.show()
```
